# preprocessamento dos dados

```python
# analise dos atributos
# transformação dos atributos categoricos em numericos para os algoritmos
# atenção na transformação para ordem de importancia usar dummy variables
colaboradores_df.info()
```

```python
X_categoricas = colaboradores_df[['BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus']]
X_categoricas
```

```python
# criação das dummy variables com as variaveis categoricas
from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder()
X_categoricas = onehotencoder.fit_transform(X_categoricas).toarray()
X_categoricas.shape
```

```python
# tipo dos dados X_categorias
# numpy array
type(X_categoricas)
```

```python
# conversão para dataframe novamente
X_categoricas = pd.DataFrame(X_categoricas)
X_categoricas
```

```python
# dataframe com atributos numericos
X_numericas = colaboradores_df[['Age', 'DailyRate', 'DistanceFromHome',	'Education', 'EnvironmentSatisfaction', 'HourlyRate', 'JobInvolvement',	'JobLevel',	'JobSatisfaction',	'MonthlyIncome',	'MonthlyRate',	'NumCompaniesWorked',	'OverTime',	'PercentSalaryHike', 'PerformanceRating',	'RelationshipSatisfaction',	'StockOptionLevel',	'TotalWorkingYears'	,'TrainingTimesLastYear'	, 'WorkLifeBalance',	'YearsAtCompany'	,'YearsInCurrentRole', 'YearsSinceLastPromotion',	'YearsWithCurrManager']]
X_numericas
```

```python
# concatenação dos dataframes com todos os atributos numericos
X_colaboradores = pd.concat([X_categoricas, X_numericas], axis=1)
X_colaboradores
```

```python
# normalização dos valores para os atributos com valores maiores para não ter pesos diferentes no algoritmo
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = scaler.fit_transform(X_colaboradores)
```

```python
# visualização da varivel
X
```

```python
# variavel alvo
y = colaboradores_df['Attrition']
y
```

```python
# separação da base de dados em treino e teste
from sklearn.model_selection import train_test_split
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.25)
```

```python
# tamanho dos dados após divisão
X_treino.shape, y_treino.shape
```

```python
X_teste.shape, y_teste.shape
```

# Regressão logística

```python
# importação da biblioteca de regressão logistica
from sklearn.linear_model import LogisticRegression
```

```python
# treinamento dos dados
RLmodel = LogisticRegression()
RLmodel.fit(X_treino, y_treino)
```

```python
# teste de previsões
RLprev = RLmodel.predict(X_teste)
RLprev
```

```python
# valores corretos para comparação
y_teste
```

```python
# calculo do accuracy
from sklearn.metrics import accuracy_score
```

```python
# uso dos valores corretos e das previsões
RLa = accuracy_score(y_teste, RLprev)
print(f'Accuracy de {RLa:2.2%} ')
```

```python
# 83.9% dos funcionarios na base de dados ficam na empresa #celula19
# importante analisar outros indicadores
# matrix de confusão
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_teste, RLprev)
cm
```

```python
# visualização grafica
# classe 0 não saem da empresa 
# calsse 1 saem da empresa
sns.heatmap(cm, annot=True);
```

```python
# precision 
# indica o percentual de chances de que a previsão indique corretamente quem vai sair
RLp = cm[1][1] / (cm[1][1] + cm[0][1])
print(f'precision de {RLp:2.2%} ')
```

```python
# recall
# o modelo indica corretamente 38.98% das pessoas que vão sair da empresa
# valor baixo para o estudo de caso
RLr = cm[1][1] / (cm[1][1] + cm[1][0])
print(f'recall de {RLr:2.2%} ')
```

```python
# usando bibliotecas sklearn
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report 
```

```python
precision_score(y_teste, RLprev)
```

```python
recall_score(y_teste, RLprev)
```

```python
# média de precisão das duas classes
f1_score(y_teste, RLprev, average='macro')
```

```python
print(classification_report(y_teste, RLprev))
```

# Redes Neurais

```python
# importação da biblioteca para redes neurais
import tensorflow as tf
```

```python
# formula para escolher quantidade de neuronios
# (quantidade de atributos + quantidade de saidas) dividido 2
(50 + 1) / 2
```

```python
# criação do modelo 
RN_model = tf.keras.models.Sequential()
RN_model.add(tf.keras.layers.Dense(units = 25, activation='relu', input_shape=(50,))) #relu valores menores que zero retorna zero
RN_model.add(tf.keras.layers.Dense(units = 25, activation = 'relu'))
RN_model.add(tf.keras.layers.Dense(units = 25, activation = 'relu'))
RN_model.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))
```

```python
# verificação do modelo
RN_model.summary()
```

```python
RN_model.compile(optimizer='Adam', loss='binary_crossentropy', metrics = ['accuracy'])
```

```python
RN_model.fit(X_treino, y_treino, epochs=200)
```

```python
RNprev = RN_model.predict(X_teste)
RNprev
```

```python
# transformação dos valores em classes pois a função sigmoide retorna valores entre 0 e 1
RNprev = (RNprev >= 0.5)
RNprev
```

```python
cm = confusion_matrix(y_teste, RNprev)
cm
```

```python
sns.heatmap(cm, annot=True);
```

```python
print(classification_report(y_teste, RNprev))
```
